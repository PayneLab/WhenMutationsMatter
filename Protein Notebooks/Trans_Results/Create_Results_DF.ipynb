# Combine Results of All Significant Trans Comparisons

```python
import pandas as pd
```

```python
ARID1A = pd.read_csv('ARID1A_Trans_Results.csv')
KRAS = pd.read_csv('KRAS_Trans_Results.csv')
FSIP2 = pd.read_csv('FSIP2_Trans_Results.csv')
MUC16 = pd.read_csv('MUC16_Trans_Results.csv')
MUC17 = pd.read_csv('MUC17_Trans_Results.csv')
MUC4 = pd.read_csv('MUC4_Trans_Results.csv')
MUC5B = pd.read_csv('MUC5B_Trans_Results.csv')
PIK3CA = pd.read_csv('PIK3CA_Trans_Results.csv')
WDFY4 = pd.read_csv('WDFY4_Trans_Results.csv')
```

```python
Trans_Results = pd.concat([ARID1A,KRAS,FSIP2,MUC16,MUC17,MUC4,MUC5B,PIK3CA,WDFY4], ignore_index=True)
```

```python
Trans_Results
```

```python
Trans_Results.to_csv(path_or_buf='All_Trans_Results.csv', index=False)
```
