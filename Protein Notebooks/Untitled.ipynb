```python
def get_frequently_mutated(gene_name):
    #import cptac
    #ov = cptac.Ovarian()
    
    protdf = ov.append_mutations_to_omics(mutation_genes = gene_name, omics_df_name = "proteomics", omics_genes = gene_name)
    protdf = protdf.loc[protdf['Sample_Status'] == "Tumor"]
    number_wt = 0
    for ind, row in protdf.iterrows():
        if len(row[gene_name+'_Mutation']) == 1 and "Wildtype_Tumor" in row[gene_name+"_Mutation"]:
            number_wt += 1
        
        if "Silent" in row[gene_name+"_Mutation"]:
            number_wt += 1
            
            
    total = len(protdf)
    
    percentage = 1 - (number_wt/total)
    return percentage
    
    
    
    
```
