```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0,2*np.pi,2000)
plot_me = 1/2*np.exp(1j*theta)-1/4*np.exp(2j*theta)
plot_me2 = 1/np.sqrt(3)*np.exp(1j*theta)-1/(3*np.sqrt(3))*np.exp(3j*theta)
```

```python
plt.scatter(np.real(plot_me), np.imag(plot_me))
plt.title('Scatter plot pythonspot.com')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(np.real(plot_me2), np.imag(plot_me2))
plt.title('Scatter plot pythonspot.com')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
