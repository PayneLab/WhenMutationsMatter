# GBM 

```python

df = pd.read_csv('GBM_EGFR_all.csv')
df = df.drop(['Unnamed: 0'], axis=1)
df
```

# Kidney

```python
df2 = pd.read_csv('Kidney_EGFR_all.csv')
```

```python
df2 = df2.drop(['Unnamed: 0'], axis=1)
df2
```

```python
Gbm_kidney = pd.merge(df, df2, on="Comparison", how = "outer")
Gbm_kidney = Gbm_kidney.rename(columns={"Slope_x": "Slope_Gbm", "R_squared_x": "R_squared_gbm","P_value_x": "P_value_Gbm" })
Gbm_kidney = Gbm_kidney.rename(columns={"Slope_y": "Slope_kidney", "R_squared_y": "R_squared_kidney","P_value_y": "P_value_kidney" })
#Gbm_kidney.to_csv("GBM_Kidney_EGFR_all.csv")
Gbm_kidney
```

```python
df3 = pd.read_csv("Endo_EGFR.csv")
df3 = df3.drop(['Unnamed: 0'], axis=1)
df3
```
