```python
def buildQuery(queryoption="HEADER", queryBy="teacherName", schoolName="", schoolID="", query=""):
    url = 'https://www.ratemyprofessors.com/search.jsp?'
    url += "queryoption=" + queryoption + "&"
    url += "queryBy=" + queryBy + "&"
    url += "schoolName=" + "+".join(schoolName.split(" ")) + "&"
    
    if (queryBy == "teacherName"):
        url += "schoolID" + schoolID + "&"
        #url += "query=" + "+".join(query.split(" "))
        
    if query != "":
        url += "query=" + "+".join(query.split(" "))
        
    #elif (queryBy == "schoolName"):
    #    url += "query=" + "+".join(query.split(" "))
        
    return(url)

q = buildQuery(schoolName="University of Utah", schoolID="1606")
```
